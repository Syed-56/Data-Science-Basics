```python
import pandas as pd
import numpy as np
url = "https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
tables = pd.read_html(url)

print(f"Total tables found: {len(tables)}\n")

for i, table in enumerate(tables):
    print(f"Table {i} shape: {table.shape}")
    print(table.head(), "\n")
```

```python
df = tables[3]
df.columns = range(df.shape[1]) #replaces column header with numbers
df = df[[0,2]]    #now our dataframe has only country name and estimated gdp by IMF
df = df.iloc[1:11,:]    #1:11 means row 1-10 and : means all columns
df.columns = ['Country','GDP (Million USD)']
print(df)
```

```python
df["GDP (Million USD)"] = df["GDP (Million USD)"].astype(int)
df["GDP (Million USD)"] = df["GDP (Million USD)"]/1000
df["GDP (Million USD)"] = np.round(df["GDP (Million USD)"],2)
df.rename(columns = {'GDP (Million USD)' : 'GDP (Billion USD)'})
```

```python
df.to_csv("GDP_extraction.csv")
with open("GDP_extraction.csv", "r") as file:
    print(file.read())
```
